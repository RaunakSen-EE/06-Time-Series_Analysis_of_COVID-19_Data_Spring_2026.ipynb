### Question 1: Import Libraries and Load Data (5 Marks)

Import `pandas` as `pd` and load the COVID-19 dataset from the provided data link into a DataFrame called `df_covid`. Then, display the first 5 rows of this DataFrame.

**Hint:** Use `pd.read_csv()` to load it and the `.head()` method to display the initial rows.

**Expected Output:** A table showing the first 5 rows of the raw COVID-19 dataset.

```python
# Write your answer here
import pandas as pd
df_covid = pd.read_csv("/content/WHO-COVID-19-global-daily-data.csv")
df_covid.head()
```

### Question 2: Subset the DataFrame (5 Marks)

Create a new DataFrame called `df_covid_subset` that contains only the following columns from the original `df_covid`: `Date_reported`, `Country`, `WHO_region`, `New_cases`, `New_deaths`. Display the first 5 rows of the new subset.

**Hint:** You can select multiple columns by passing a list of column names to the DataFrame, like `df[['col1', 'col2']]`.

**Expected Output:** A table showing the first 5 rows with only the 5 specified columns.

```python
# Write your answer here
df_covid_subset = df_covid[['Date_reported', 'Country', 'WHO_region', 'New_cases', 'New_deaths']].copy()
df_covid_subset.head()
```

### Question 3: Data Type Conversion (5 Marks)

The `Date_reported` column in `df_covid_subset` is currently an object (string). Convert this column to a proper datetime format. Then, print the `info()` of the DataFrame to confirm the change.

**Hint:** Use `pd.to_datetime(df['column_name'])` to perform the conversion. Use `df.info()` to check the data types.

**Expected Output:** A summary from `.info()` showing that `Date_reported` is now of type `datetime64[ns]`.

```python
# Write your answer here
df_covid_subset['Date_reported'] = pd.to_datetime(df_covid_subset['Date_reported'])
df_covid_subset.info()
```

### Question 4: Filter by Date Range (10 Marks)

Create a new DataFrame called `df_covid_trimmed` by filtering `df_covid_subset` to include only the dates from '2020-03-01' to '2023-08-31', inclusive. Print the shape of the new trimmed DataFrame.

**Hint:** You can filter a datetime column using string comparisons, e.g., `df[(df['Date'] >= 'start_date') & (df['Date'] <= 'end_date')]`.

**Expected Output:** A tuple representing the shape of the filtered DataFrame (rows, columns).

```python
# Write your answer here
df_covid_trimmed = df_covid_subset[
    (df_covid_subset['Date_reported'] >= '2020-03-01') &
    (df_covid_subset['Date_reported'] <= '2023-08-31')].copy()

df_covid_trimmed.shape
```

### Question 5: Find Top 5 Affected Countries (10 Marks)

From the original `df_covid` DataFrame, find the 5 countries with the highest total `Cumulative_cases`. Group the data by `Country`, find the maximum `Cumulative_cases` for each, sort the results, and get the top 5. Print the resulting Series.

**Hint:** Use a chain of methods: `.groupby('Country')['Cumulative_cases'].max().sort_values(ascending=False).head(5)`.

**Expected Output:** A Pandas Series showing the top 5 countries as the index and their max cumulative cases as the values.

```python
# Write your answer here
top_5_countries = (df_covid.groupby('Country')['Cumulative_cases'].max().sort_values(ascending=False).head(5))
top_5_countries
```

### Question 6: Calculate Daily Global Cases (10 Marks)

Using the `df_covid_trimmed` DataFrame, calculate the total number of `New_cases` for each date across all countries. Store this in a new DataFrame called `daily_global_cases` and print its head.

**Hint:** Use `.groupby('Date_reported')['New_cases'].sum()`.

**Expected Output:** A Series showing the first few dates and the corresponding total new cases for that day.

```python
# Write your answer here
daily_global_cases = (df_covid_trimmed.groupby('Date_reported')['New_cases'].sum())
daily_global_cases.head()
```

### Question 7: Aggregate Data by Quarter (10 Marks)

Aggregate the `New_cases` and `New_deaths` from `df_covid_trimmed` on a quarterly basis. First, create a 'Quarter' column, then group by this column and sum the cases and deaths. Display the head of the resulting DataFrame.

**Hint:** Create the quarter column with `df['Date_reported'].dt.to_period('Q')`. Then, group by this new column and use `.agg({'New_cases': 'sum', 'New_deaths': 'sum'})`.

**Expected Output:** A DataFrame with 'Quarter' as the index and columns for 'New_cases' and 'New_deaths'.

```python
# Write your answer here
df_covid_trimmed['Quarter'] = df_covid_trimmed['Date_reported'].dt.to_period('Q')
quarterly_data = (df_covid_trimmed.groupby('Quarter').agg({'New_cases': 'sum','New_deaths': 'sum'}))
quarterly_data.head()
```

### Question 8: Total Cases by WHO Region (15 Marks)

Calculate the total sum of `New_cases` for each `WHO_region` in the `df_covid_trimmed` DataFrame. Display the results sorted in descending order.

**Hint:** Group by `WHO_region`, get the sum of `New_cases`, and then use `.sort_values(ascending=False)`.

**Expected Output:** A Pandas Series showing each WHO region and their total new cases, sorted from highest to lowest.

```python
# Write your answer here
cases_by_region = (df_covid_trimmed.groupby('WHO_region')['New_cases'].sum().sort_values(ascending=False))
cases_by_region
```

### Question 9: Find the Day with the Highest Number of New Cases (15 Marks)

Using the `daily_global_cases` Series you created earlier, find the date on which the highest number of new cases was reported globally. Print the date and the corresponding number of cases.

**Hint:** The `.idxmax()` method on a Series will return the index (in this case, the date) of the maximum value. You can use `.max()` to get the maximum value itself.

**Expected Output:** A printout of the date with the highest cases and the number of cases on that day.

```python
# Write your answer here
max_cases_date = daily_global_cases.idxmax()
max_cases_value = daily_global_cases.max()
print("Date with highest global new cases:", max_cases_date)
print("Number of cases on that day:", max_cases_value)
```

### Question 10: Pivot Table for Monthly Cases by Region (15 Marks)

Create a pivot table to show the total `New_cases` for each `WHO_region` aggregated by month. First, create a 'Month_Year' column. Then, create the pivot table with `WHO_region` as the index, `Month_Year` as columns, and the sum of `New_cases` as values. Display the resulting pivot table.

**Hint:** Create the month column with `df['Date_reported'].dt.to_period('M')`. Use `pd.pivot_table()` with `index`, `columns`, `values`, and `aggfunc='sum'`.

**Expected Output:** A table with regions as rows, month-years as columns, and aggregated new cases as the cell values.

```python
# Write your answer here
df_covid_trimmed.loc[:, 'Month_Year'] = (df_covid_trimmed['Date_reported'].dt.to_period('M'))
pivot_monthly_cases = pd.pivot_table(df_covid_trimmed,index='WHO_region',columns='Month_Year',values='New_cases',aggfunc='sum')
pivot_monthly_cases
```
